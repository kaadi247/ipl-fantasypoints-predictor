```python
valid_match_ids = set(matches['id'])
deliveries = deliveries[deliveries['match_id'].isin(valid_match_ids)]
print(f"Deliveries after match filter: {len(deliveries)}")
```

```python
# Step 1: Aggregate batting stats per player per match
batting = deliveries.groupby(['match_id', 'batter']).agg(
    runs=('batsman_runs', 'sum'),
    fours=('batsman_runs', lambda x: (x == 4).sum()),
    sixes=('batsman_runs', lambda x: (x == 6).sum()),
).reset_index()

# Step 2: Find dismissed batters to check for ducks
dismissed = deliveries[deliveries['is_wicket'] == 1][['match_id', 'player_dismissed']].copy()
dismissed.columns = ['match_id', 'batter']
dismissed['got_out'] = 1

batting = batting.merge(dismissed, on=['match_id', 'batter'], how='left')
batting['got_out'] = batting['got_out'].fillna(0)
batting['duck'] = ((batting['runs'] == 0) & (batting['got_out'] == 1)).astype(int)

# Step 3: Calculate batting fantasy points
batting['bat_pts'] = (
    batting['runs'] * 0.5 +
    batting['fours'] * 1 +
    batting['sixes'] * 2 +
    (batting['runs'] >= 25).astype(int) * 4 +
    (batting['runs'] >= 50).astype(int) * 8 +
    (batting['runs'] >= 75).astype(int) * 8 +
    (batting['runs'] >= 100).astype(int) * 16 -
    batting['duck'] * 2
)

print(batting[['batter', 'runs', 'fours', 'sixes', 'duck', 'bat_pts']].head(10))
print(f"\nTotal batting entries: {len(batting)}")
print(f"Avg batting pts: {batting['bat_pts'].mean():.2f}")
```

```python
# Step 1: Wickets - run outs credit the fielder, not the bowler
bowl_deliveries = deliveries[
    (deliveries['is_wicket'] == 1) & 
    (~deliveries['dismissal_kind'].isin(['run out', 'retired hurt', 'retired out']))
]

bowling = bowl_deliveries.groupby(['match_id', 'bowler']).agg(
    wickets=('is_wicket', 'sum'),
    lbw_bowled=('dismissal_kind', lambda x: x.isin(['lbw', 'bowled']).sum())
).reset_index()

# Step 2: Maiden overs
over_runs = deliveries.groupby(['match_id', 'bowler', 'over'])['total_runs'].sum().reset_index()
maidens = over_runs[over_runs['total_runs'] == 0].groupby(
    ['match_id', 'bowler']
).size().reset_index(name='maidens')

bowling = bowling.merge(maidens, on=['match_id', 'bowler'], how='left')
bowling['maidens'] = bowling['maidens'].fillna(0)

# Step 3: Calculate bowling fantasy points
bowling['bowl_pts'] = (
    bowling['wickets'] * 25 +
    bowling['lbw_bowled'] * 8 +
    (bowling['wickets'] >= 3).astype(int) * 4 +
    (bowling['wickets'] >= 4).astype(int) * 8 +
    (bowling['wickets'] >= 5).astype(int) * 16 +
    bowling['maidens'] * 12
)

print(bowling[['bowler', 'wickets', 'lbw_bowled', 'maidens', 'bowl_pts']].head(10))
print(f"\nTotal bowling entries: {len(bowling)}")
print(f"Avg bowling pts: {bowling['bowl_pts'].mean():.2f}")
```

```python
# Catches (fielder column has the catcher's name)
catches = deliveries[deliveries['dismissal_kind'] == 'caught'].dropna(subset=['fielder'])
catches_grp = catches.groupby(['match_id', 'fielder']).size().reset_index(name='catches')
catches_grp.rename(columns={'fielder': 'player'}, inplace=True)

# Caught and bowled - bowler gets the credit
cab = deliveries[deliveries['dismissal_kind'] == 'caught and bowled'].copy()
cab_grp = cab.groupby(['match_id', 'bowler']).size().reset_index(name='cab_catches')
cab_grp.rename(columns={'bowler': 'player'}, inplace=True)

# Stumpings
stumpings = deliveries[deliveries['dismissal_kind'] == 'stumped'].dropna(subset=['fielder'])
stumpings_grp = stumpings.groupby(['match_id', 'fielder']).size().reset_index(name='stumpings')
stumpings_grp.rename(columns={'fielder': 'player'}, inplace=True)

# Run outs
run_outs = deliveries[deliveries['dismissal_kind'] == 'run out'].dropna(subset=['fielder'])
run_outs_grp = run_outs.groupby(['match_id', 'fielder']).size().reset_index(name='run_outs')
run_outs_grp.rename(columns={'fielder': 'player'}, inplace=True)

# Merge all fielding components
fielding = catches_grp.merge(cab_grp, on=['match_id', 'player'], how='outer')
fielding = fielding.merge(stumpings_grp, on=['match_id', 'player'], how='outer')
fielding = fielding.merge(run_outs_grp, on=['match_id', 'player'], how='outer')
fielding = fielding.fillna(0)

fielding['field_pts'] = (
    fielding['catches'] * 8 +
    fielding['cab_catches'] * 8 +
    fielding['stumpings'] * 12 +
    fielding['run_outs'] * 6
)

print(fielding.head(10))
print(f"\nTotal fielding entries: {len(fielding)}")
```

```python
# Standardise column names for merging
bat_merge = batting[['match_id', 'batter', 'bat_pts']].rename(columns={'batter': 'player'})
bowl_merge = bowling[['match_id', 'bowler', 'bowl_pts']].rename(columns={'bowler': 'player'})
field_merge = fielding[['match_id', 'player', 'field_pts']]

# Outer merge — a player might bat but not bowl, or bowl but not bat
fantasy = bat_merge.merge(bowl_merge, on=['match_id', 'player'], how='outer')
fantasy = fantasy.merge(field_merge, on=['match_id', 'player'], how='outer')
fantasy = fantasy.fillna(0)

fantasy['fantasy_points'] = fantasy['bat_pts'] + fantasy['bowl_pts'] + fantasy['field_pts']

print(f"Total player-match entries: {len(fantasy)}")
print(f"\nFantasy points distribution:")
print(fantasy['fantasy_points'].describe())
print(f"\nTop 10 individual performances:")
print(fantasy.sort_values('fantasy_points', ascending=False).head(10))
```

```python
# Bring in match context
match_context = matches[['id', 'season', 'date', 'venue', 'team1', 'team2', 
                          'toss_winner', 'toss_decision', 'winner']].copy()

fantasy = fantasy.merge(
    match_context, 
    left_on='match_id', 
    right_on='id', 
    how='left'
).drop(columns='id')

# Add a column: which team was this player playing for?
# Get player-team mapping from deliveries
batting_teams = deliveries[['match_id', 'batter', 'batting_team']].rename(
    columns={'batter': 'player', 'batting_team': 'team'}
)
bowling_teams = deliveries[['match_id', 'bowler', 'bowling_team']].rename(
    columns={'bowler': 'player', 'bowling_team': 'team'}
)
player_teams = pd.concat([batting_teams, bowling_teams]).drop_duplicates()

fantasy = fantasy.merge(player_teams, on=['match_id', 'player'], how='left')

# Did the player's team win?
fantasy['won'] = (fantasy['team'] == fantasy['winner']).astype(int)

# Save to disk
fantasy.to_csv('data/fantasy_points.csv', index=False)

print(f"Saved {len(fantasy)} player-match records")
print(f"\nColumns: {list(fantasy.columns)}")
print(f"\nSample:")
print(fantasy.head(5).to_string())
```
